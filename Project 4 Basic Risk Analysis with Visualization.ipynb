# Step 1: Simulated Data

```python
import pandas as pd
import numpy as np

# Simulate daily returns for one year
np.random.seed(42)
dates = pd.date_range(start='2022-01-01', periods=365)
returns = np.random.normal(loc=0.0005, scale=0.02, size=len(dates))

# Create DataFrame
df = pd.DataFrame({'Date': dates, 'Returns': returns})
```

# Step 2: Visualize Returns Distribution

```python
# Creating a histogram to visualize the distribution of returns.
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for the plot
sns.set(style='whitegrid')

# Create a histogram of Returns
plt.figure(figsize=(10, 6))
sns.histplot(df['Returns'], bins=30, kde=True, color='blue')
plt.title('Distribution of Daily Returns')
plt.xlabel('Returns')
plt.ylabel('Frequency')
plt.axvline(x=np.percentile(df['Returns'], 5), color='red', linestyle='--', label='95% VaR')
plt.axvline(x=np.mean(df['Returns']), color='green', linestyle='-.', label='Mean Return')
plt.legend()
plt.show()
```
